```python
netflix = pd.read_csv("netflix.csv", parse_dates=["date_added"])
```

```python
netflix.nunique()
```

```python
netflix["type"] = netflix["type"].astype("category")
```

```python
netflix.info()
```

```python
is_Limitless = netflix["title"] == "Limitless"
netflix[is_Limitless]
```

```python
direct_by_RR = netflix["director"] == "Robert Rodriguez"
is_movie = netflix["type"] == "Movie"
netflix[direct_by_RR & is_movie]
```

```python
regidate_190731 = netflix["date_added"] == "2019-07-31"
direct_by_RA = netflix["director"] == "Robert Altman"
netflix[regidate_190731 | direct_by_RA]
```

```python
directors = ["Orson Welles", "Aditya Kripalani", "Sam Raimi"]
target_directors = netflix["director"].isin(directors)
netflix[target_directors]
```

```python
netflix[netflix["date_added"].between("2019-05-01", "2019-06-01")]
```

```python
netflix = netflix.dropna(subset=["director"])
netflix
```

```python
netflix.drop_duplicates(subset=["date_added"], keep=False)
```
