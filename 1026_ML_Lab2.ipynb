# Breast Cancer Classification

## 목적

scikit-learn에서 제공하는 breast_cancer 데이터셋, Decision Trees, Support Vector Machines (SVM)을 사용하여 종양이 악성인지 양성인지를 예측하는 분류 모델을 구축합니다.

## 절차

1. scikit-learn에서 breast_cancer 데이터셋을 로드합니다. (load_breast_cancer)
2. 데이터셋을 탐색하고 그 특성 및 대상 변수를 이해합니다. (matplotlib)
3. 데이터셋을 훈련 세트와 테스트 세트로 분할합니다. (train_test_split)
4. 두 분류 모델을 구현합니다:
> * 결정 트리 분류기 (DecisionTreeClassifier ...)
> * 서포트 벡터 머신 (SVM) 분류기 (SVC ...)

5. 훈련 세트를 사용하여 각 모델을 훈련합니다. (fit)
6. 테스트 세트에서 각 모델의 성능을 평가합니다. (score)
7. 두 모델의 성능을 비교해 봅니다.

```python
import numpy as np
import matplotlib.pyplot as plt
!pip install koreanize-matplotlib
import koreanize_matplotlib

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
cancer = load_breast_cancer()
```

```python
feature_names = cancer.feature_names
X = cancer.data

plt.scatter(X[:, 0], X[:, 1], c=cancer.target, cmap=plt.cm.coolwarm)
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("암")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
tree = DecisionTreeClassifier(random_state=0)
```

```python
svm = SVC(random_state=0)
```

```python
tree.fit(X_train, y_train)
```

```python
svm.fit(X_train, y_train)
```

```python
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
print("훈련 세트 정확도: {:.3f}".format(svm.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(svm.score(X_test, y_test)))
```
