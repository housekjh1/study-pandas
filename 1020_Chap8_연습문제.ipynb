# 재구성화 연습문제 풀이

```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("used_cars.csv")
cars
```

```python
cars.pivot_table(values="Price", index="Fuel", aggfunc="sum")
```

```python
cars.pivot_table(index="Manufacturer", aggfunc="count", columns="Transmission", values="Price", margins=True)
```

```python
cars.pivot_table(index=["Year", "Fuel"], aggfunc="mean", columns="Transmission", values="Price")
```

```python
report = cars.pivot_table(index=["Year", "Fuel"], aggfunc="mean", columns="Transmission", values="Price").stack()
```
