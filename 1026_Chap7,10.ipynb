```python
customers["Address"].str.split(",", expand=True)
```

```python
customers["Street"] = customers["Address"].str.split(",", expand=True)[0]
customers["City"] = customers["Address"].str.split(",", expand=True)[1]
customers["State"] = customers["Address"].str.split(",", expand=True)[2]
customers["Zip"] = customers["Address"].str.split(",", expand=True)[3]
customers
```

```python
new_cols = ["Street", "City", "State", "Zip"]
customers[new_cols] = customers["Address"].str.split(",", expand=True)
customers
```

```python
import FinanceDataReader as fdr
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = fdr.DataReader("AAPL", "2023")
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df.index, y=df["Close"])
plt.show()
```

```python
df_month = df.resample("BM").mean()
df_month.head()
```

```python
df_month['RTN']=df_month['Close'].pct_change()*100
df_month.head()
```

```python
df_month['MA']=df_month['Close'].rolling(2).mean()
df_month.head()
```
