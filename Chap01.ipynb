### 1.3.4 하나 이상의 기준으로 열 필터링

유니버셜(Universal) 스튜디오에서 제작한 영화만 찾고 싶다면?

```python
movies[movies["Studio"]=="Universal"]
```

필터링 조건을 변수에 저장하면 가독성을 높일 수 있음

```python
released_by_universal = movies["Studio"]=="Universal"
movies[released_by_universal]
```

유니버설에서 제작하고, 2015년에 개봉한 모든 영화를 찾으면?

```python
released_by_universal = movies["Studio"]=="Universal"
released_in_2015 = movies["Year"]==2015
movies[released_by_universal&released_in_2015]
```

```python
movies[(movies["Studio"]=="Universal")&(movies["Year"]==2015)]
```

유니버설에서 제작하거나, 2015년에 개봉한 영화는?

```python
movies[(movies["Studio"]=="Universal")|(movies["Year"]==2015)]
```

특정 값보다 작거나 큰 열 값을 필터링하기
- 1975년 이전에 개봉한 영화 찾기

```python
before_1975 = movies["Year"] < 1975
movies[before_1975]
```

특정 범위에 해당하는 영화를 찾는 경우
- 1983년에서 1986년 사이에 개봉한 영화는?

```python
between_1983_1986 = movies["Year"].between(1983,1986)
movies[between_1983_1986]
```

인덱스를 이용한 행 필터링하기
- "dark" 단어가 영화 제목에 들어가는 영화 찾기
- pandas에서 제공하는 메서드인 str을 이용하여 영화 제목을 소문자로 바꾸고 "dark"단어를 포함한 행 가져오기

```python
d = movies.index.str.lower().str.contains("dark")
movies[d]
```

1. 영화 제목이 인덱스이므로 인덱스 값을 가져온다.
2. 인덱스 문자열 값에 대소문자가 섞여 있으므로 소문자로 변환한다.
3. "dark"라는 글자가 포함된 인덱스만 True로 나오는 Array를 반환받는다.
4. True인 위치의 행만 추출한다.

### 1.3.5 데이터 그룹화

어떤 제작사의 총수익이 가장 높은지
- Gross 열의 총합

```python
movies.dtypes
```

<문제점>
- "Gross"의 제작사별 총합을 계산해야하는데 "Gross"열은 숫자가 아닌 텍스트
 - 텍스트 -> 숫자 형변화

- 형변환하는 과정에서 문제가 발생
 - '&', ","를 먼저 필터링 필요

```python
movies["Gross"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
```

object -> float64 형변환

```python
movies["Gross"].str.replace("$", "", regex=False).str.replace(",", "", regex=False).astype(float)
```
