```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fortune = pd.read_csv("fortune1000.csv")
fortune.head(10)
```

```python
fortune.info()
```

```python
fortune.nunique()
```

```python
is_retailing = fortune["Sector"] == "Retailing"
retail_companies = fortune[is_retailing]
retail_companies.head()
```

```python
retail_companies["Revenues"].mean()
```

```python
sectors = fortune.groupby("Sector")
sectors
```

```python
len(sectors) == fortune["Sector"].nunique()
```

```python
sectors.size()
```

```python
sectors.groups
```

```python
fortune.loc[26, "Sector"]
```

```python
sectors.first()
```

```python
sectors.last()
```

```python
sectors.nth(10)
```

```python
sectors.head()
```

```python
sectors.get_group("Energy").head()
```

```python
fortune.nlargest(n=5, columns="Profits")
```

```python
def get_largest_row(df):
  return df.nlargest(1, "Revenues")
sectors.apply(get_largest_row).head()
```

```python
cereals = pd.read_csv("cereals.csv")
cereals.head()
```

```python
cereals.info()
```

```python
manufacturers = cereals.groupby("Manufacturer")
```

```python
manufacturers.size()
```

```python
len(manufacturers)
```

```python
manufacturers.get_group("Nabisco")
```

```python
manufacturers.mean()
```

```python
manufacturers["Sugars"].max()
```

```python
manufacturers["Fiber"].min()
```

```python
def smallest_sugar_row(df):
  return df.nsmallest(1, "Sugars")

manufacturers.apply(smallest_sugar_row)
```
